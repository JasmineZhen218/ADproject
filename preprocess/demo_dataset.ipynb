### Show samples of AD dataset
* You will see 5 samples of the AD disease dataset
* Each sample is a ndarray with shape (132,132,3)
* The first 3 channels corresponds to region cropped from base (original pathology image)
* The last channel corresponds to pixel-wise labelling. 1 represents biomarker, 0 represent background

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
for i in range(5):
    ID=i+1
    sample_name="samples/"+str(ID)+'.npy'
    sample=np.load(sample_name)
    print(sample.shape)
    base=sample[:,:,0:3]
    label=sample[:,:,3]
    f,(ax1,ax2)=plt.subplots(1,2)
    ax1.imshow(base)
    ax2.imshow(label)
```
