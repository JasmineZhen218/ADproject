### Show samples of AD dataset
* There are 496 samples in https://github.com/JasmineZhen218/ADproject/blob/master/preprocess/asset/Samples
* Each sample is a ndarray with shape (132,132,4)
* The first 3 channels corresponds to region cropped from base (original pathology image)
* The last channel corresponds to pixel-wise labelling. 1 represents biomarker, 0 represent background

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
for i in range(10):
    ID=random.choice(np.arange(1,496))
    sample_name="asset/Samples/"+str(ID)+'.npy'
    sample=np.load(sample_name)
    print(sample.shape)
    base=sample[:,:,0:3]
    label=sample[:,:,3]
    f,(ax1,ax2)=plt.subplots(1,2)
    ax1.imshow(base)
    ax2.imshow(label)
    ax1.set_title("Base_"+str(ID))
    ax2.set_title("Label_"+str(ID))
```
