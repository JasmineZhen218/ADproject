# Visualize pathology image and probability map
* The pathology image and probability map are in []
* The probability map is the result of applying Pre-trained UNET model to pathology image
* The examples below are a pathology images stained using antibody of amyloid plauque and its corresponding probability map got using `unet_amyloid.pth`. 

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

### The whole image

```python
Base=mpimg.imread("asset/Base.tif")
plt.imshow(Base)
plt.title("The original pathology image")
```

```python
Prob_map=np.load("asset/Prob_map.npy")
plt.imshow(Prob_map)
plt.title("The Probability map")
```

### Crop a small region and see details

```python
Base_crop=Base[2000:4000,10000:12000]
Prob_map_crop=Prob_map[2000:4000,10000:12000]
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,20))
ax1.imshow(Base_crop)
ax1.set_title("The original pathology image")

ax2.imshow(Prob_map_crop)
ax2.set_title("The Probability map")
```
